```
LONDON_HOUSE_AREA_BY_BOROUGH = 'floorarea-data-borough.xlsx'
```

```
area_by_borough = pd.read_excel(os.path.join(BASE_PATH, LONDON_HOUSE_AREA_BY_BOROUGH))
```

```
average_area = area_by_borough[['Code', 'Name', 'Av floorspace of houses (m2)', 'Av floorspace of flats (m2)', 'Av floorspace of All properties (m2)']]
```

```
average_area = average_area.iloc[0:33, :]
average_area['District'] = average_area.Name.apply(lambda x : x.upper())
average_area['District'] = average_area.District.apply(lambda x : x[:-1])
```

```
average_area
```

```
sorted(london_house_transactions_and_prices['District'].unique())
```

```
sorted(average_area['District'].unique())
```

```
data_merged = pd.merge(london_house_transactions_and_prices, average_area, on=["District", 'District'], how = 'left')
```

```
data_merged['Is_House'] = data_merged['Property_Type'].apply(lambda x: 0 if (x == 'F') else 1)
data_merged['Is_Flat'] = data_merged['Property_Type'].apply(lambda x: 1 if (x == 'F') else 0)
```

```
data_merged['Expected Floor Space'] = data_merged['Av floorspace of houses (m2)']*data_merged['Is_House']+data_merged['Av floorspace of flats (m2)']*data_merged['Is_Flat']
data_merged['Price Per Squared Meter'] = data_merged['Price']/data_merged['Expected Floor Space']
```

```
data_merged = data_merged[data_merged.PPD != 'B']
```

```
data_merged
```

```
data_merged.to_csv(os.path.join(BASE_PATH , 'London_house_transaction_price_per_sqm.csv'))
```
